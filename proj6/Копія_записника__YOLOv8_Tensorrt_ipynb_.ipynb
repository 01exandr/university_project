```python
import cv2
import os
import torch
import numpy as np

from ultralytics.utils import ops
from ultralytics.data.augment import LetterBox
from ultralytics.engine.results import Results

def get_text_color(box_color):
    text_color = (255,255,255)

    brightness = box_color[2]*0.299 + box_color[1]*0.587 + box_color[0]*0.114

    if(brightness > 180):
        text_color = (0, 0, 0)

    return text_color

def draw_box(img, detection_output, class_list, colors) :
    # Copy image, in case that we need original image for something
    out_image = img

    for run_output in detection_output :
        # Unpack
        label, con, box = run_output

        # Choose color
        box_color = colors[int(label.item())]
        # text_color = (255,255,255)
        text_color = get_text_color(box_color)
        # Get Class Name
        label = class_list[int(label.item())]
        # Draw object box
        first_half_box = (int(box[0].item()),int(box[1].item()))
        second_half_box = (int(box[2].item()),int(box[3].item()))
        cv2.rectangle(out_image, first_half_box, second_half_box, box_color, 2)
        # Create text
        text_print = '{label} {con:.2f}'.format(label = label, con = con.item())
        # Locate text position
        text_location = (int(box[0]), int(box[1] - 10 ))
        # Get size and baseline
        labelSize, baseLine = cv2.getTextSize(text_print, cv2.FONT_HERSHEY_SIMPLEX, 1, 1)

        # Draw text's background
        cv2.rectangle(out_image
                        , (int(box[0]), int(box[1] - labelSize[1] - 10 ))
                        , (int(box[0])+labelSize[0], int(box[1] + baseLine-10))
                        , box_color , cv2.FILLED)
        # Put text
        cv2.putText(out_image, text_print ,text_location
                    , cv2.FONT_HERSHEY_SIMPLEX , 1
                    , text_color, 2, cv2.LINE_AA)

    return out_image

def draw_fps(avg_fps, combined_img):
    avg_fps_str = float("{:.2f}".format(avg_fps))

    # cv2.rectangle(combined_img, (10,2), (110,50), (255,255,255), -1) # (660,110)
    cv2.putText(combined_img, "FPS: "+str(avg_fps_str), (5, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), thickness=3) #(20, 90) 3.5 6

    return combined_img

def get_name(source_path):
    name_idx = 0
    file_pos = (source_path).rfind('\\')

    if(file_pos == -1):
        file_pos = (source_path).rfind('/')

        if(file_pos == -1):
            file_pos = 0

    name_idx = file_pos + 1

    name = source_path[name_idx:]

    return name

def get_save_path(file_name, folder_name):
    path = "result"
    save_path = os.path.join(path, folder_name)

    exists = os.path.exists(save_path)

    if(not exists):
        os.makedirs(save_path)

    save_path = os.path.join(save_path, file_name)

    return save_path

def preprocess(img):
    # LetterBox
    im = LetterBox((640, 640), False)(image=img)
    im = im.transpose((2, 0, 1))[::-1]  # HWC to CHW, BGR to RGB
    im = np.ascontiguousarray(im)

    img = torch.from_numpy(im).to(torch.device('cuda:0'))
    img = img.half()  # uint8 to fp16/32
    img /= 255  # 0 - 255 to 0.0 - 1.0

    if len(img.shape) == 3:
        img = img[None]  # expand for batch dim

    return img

def postprocess(preds, img, orig_img, names, source_path):
    preds = ops.non_max_suppression(preds,
                                    0.5,
                                    0.7,
                                    agnostic=False,
                                    max_det=300,
                                    classes=None)

    results = []
    for i, pred in enumerate(preds):
        orig_img = orig_img[i] if isinstance(orig_img, list) else orig_img
        shape = orig_img.shape
        pred[:, :4] = ops.scale_boxes(img.shape[2:], pred[:, :4], shape).round()
        path = source_path
        img_path = path[i] if isinstance(path, list) else path
        results.append(Results(orig_img=orig_img, path=img_path, names=names, boxes=pred))

    return results

def frame_objects_count(frame, detections, class_list, colors):
    """
    Processes a single frame to draw bounding boxes, count objects, and display the count.

    Args:
        frame (numpy.ndarray): The input video frame.
        detections (list): List of detections in the format [(cls, conf, [x1, y1, x2, y2])].
        class_list (list): List of class names corresponding to the model's class indices.
        colors (list): List of colors for bounding boxes for each class.

    Returns:
        tuple: Processed frame with drawn bounding boxes and the object count.
    """
    frame_objects_count = 0

    for cls, conf, box in detections:
        frame_objects_count += 1

        # Unpack bounding box coordinates
        x1, y1, x2, y2 = box

        # Determine the label and box color
        label = class_list[int(cls)]
        box_color = colors[int(cls)]

        # Get the text color based on box brightness
        text_color = get_text_color(box_color)

        # Draw bounding box
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), box_color, 2)

        # Prepare label with confidence score
        label_with_conf = f"{label} {conf:.2f}"

        # Draw label background and text
        text_size, _ = cv2.getTextSize(label_with_conf, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)
        text_width, text_height = text_size
        cv2.rectangle(frame, (int(x1), int(y1) - text_height - 4),
                      (int(x1) + text_width, int(y1)), box_color, -1)
        cv2.putText(frame, label_with_conf, (int(x1), int(y1) - 2),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1)

    # Display frame objects count on the frame
    cv2.putText(frame, f"Frame Objects: {frame_objects_count}", (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

    return frame, frame_objects_count
```

```python
import torch
import cv2
import random
import time
import pathlib
from ultralytics import YOLO

# import modules.utils as utils

def get_detection_result(model, frame):
    # Update object localizer
    results = model.predict(frame, imgsz=640, conf=0.5, verbose=False)
    result = results[0].cpu()

    # Get information from result
    box = result.boxes.xyxy.numpy()
    conf = result.boxes.conf.numpy()
    cls = result.boxes.cls.numpy().astype(int)

    return cls, conf, box

def detection(model_path, source, name):
    # Check File Extension
    file_extension = pathlib.Path(model_path).suffix

    # Load the Model
    model = YOLO(model_path)

    # Class Name and Colors
    label_map = model.names
    COLORS = [[random.randint(0, 255) for _ in range(3)] for _ in label_map]

    # FPS Detection
    frame_count = 0
    total_fps = 0
    avg_fps = 0

    # FPS Video
    video_cap = cv2.VideoCapture(source)
    total_frames = int(video_cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_width = int(video_cap.get(3))
    frame_height = int(video_cap.get(4))

    video_frames = []

    while video_cap.isOpened():
        ret, frame = video_cap.read()
        if not ret:
            break

        # Start Time
        start = time.time()

        # Detection
        cls, conf, box = get_detection_result(model, frame)
        detection_output = list(zip(cls, conf, box))

        # Count objects and draw bounding boxes
        frame_objects_count_value = 0  # Initialize count for this frame
        frame, frame_objects_count_value = frame_objects_count(frame, detection_output, label_map, COLORS)

        # End Time
        end = time.time()

        # Draw FPS
        frame_count += 1
        fps = 1 / (end - start)
        total_fps += fps
        avg_fps = total_fps / frame_count

        # Draw FPS on the frame
        frame = draw_fps(avg_fps, frame)

        # Append processed frame to the video output
        video_frames.append(frame)

        print(f"Frame {frame_count}/{total_frames} processed. Objects detected: {frame_objects_count_value}")

    # Save processed video
    file_name = get_name(source)
    save_path = get_save_path(file_name, name)
    out = cv2.VideoWriter(save_path, cv2.VideoWriter_fourcc(*'XVID'), int(avg_fps), (frame_width, frame_height))

    for frame in video_frames:
        out.write(frame)

    out.release()

    print(f"Video is saved in: {save_path}")
```

```python
import time
import cv2
import torch
from ultralytics import YOLO
from sahi.predict import get_sliced_prediction

# ------------------------------
# User Configurable Section
# ------------------------------
VIDEO_PATH = "/content/Cars1.mp4"
OUTPUT_PATH = "/content/output9_video.mp4"

# Specify one or two model paths:
# Supports .pt (Pytorch), .onnx (ONNX), or .engine (TensorRT)
MODEL_PATHS = [
    "/content/est.onnx"   # first model (required)
    # "D:\magl\code_jet\models\model_next\last.pt"
    # "D:\magl\code_jet\yolov8n.onnx" # second model (uncomment if you want to run two models)
]

USE_SAHI_SLICING = False        # Set to True if you want slicing for large images
SLICE_SIZE = 640
OVERLAP_RATIO = 0.2
CONF_THRESHOLD = 0.25

# ------------------------------
# Helper Classes & Functions
# ------------------------------

class YOLOv8ModelWrapper:
    """Wraps a YOLOv8 model for optional SAHI slicing."""
    def __init__(self, model_path, conf=0.25, device=None):
        self.model_path = model_path
        self.conf = conf
        self.device = device if device else ('cuda' if torch.cuda.is_available() else 'cpu')
        self.model = YOLO(model_path)
        self.model.to(self.device)

    def predict_frame(self, frame, use_slicing=False, slice_size=640, overlap_ratio=0.2):
        """Predict on a single frame. If use_slicing is True, apply SAHI slicing."""
        if use_slicing:
            # We adapt YOLOv8 model for SAHI if needed
            # Create a custom adapter for SAHI
            detection_list = self._sahi_sliced_inference(frame, slice_size, overlap_ratio)
        else:
            # Direct YOLOv8 inference
            results = self.model.predict(frame, conf=self.conf, verbose=False, device=self.device)
            detection_list = results[0].boxes.data.cpu().numpy() if len(results) > 0 else []
        return detection_list

    def _sahi_sliced_inference(self, frame, slice_size, overlap_ratio):
        # Create a temporary SAHI adapter for each call (if needed).
        # If SAHI adds direct YOLOv8 support, you can integrate it directly.
        detection_list = []

        class YOLOv8SAHIAdapter:
            def __init__(self, yolov8_wrapper):
                self.yolov8_wrapper = yolov8_wrapper

            def predict(self, image, conf_thres=0.25):
                results = self.yolov8_wrapper.model.predict(image, conf=conf_thres, verbose=False, device=self.yolov8_wrapper.device)
                predictions = results[0].boxes.data.cpu().numpy() if len(results) > 0 else []
                # Convert to SAHI format
                sahi_dets = []
                for det in predictions:
                    x1, y1, x2, y2, score, cls = det
                    sahi_dets.append({
                        "bbox": [float(x1), float(y1), float(x2 - x1), float(y2 - y1)],
                        "score": float(score),
                        "category_id": int(cls)
                    })
                return sahi_dets

        # Perform SAHI slicing
        adapter = YOLOv8SAHIAdapter(self)
        sliced_result = get_sliced_prediction(
            image=frame,
            detection_model=adapter,
            slice_size=slice_size,
            overlap_ratio=overlap_ratio,
            postprocess_match_threshold=0.5,
            verbose=0
        )
        # sliced_result.object_prediction_list are SAHI predictions
        # Convert these back to YOLOv8-like format [x1, y1, x2, y2, conf, class]
        for obj_pred in sliced_result.object_prediction_list:
            xyxy = obj_pred.bbox.to_xyxy()
            score = obj_pred.score.value
            cls = obj_pred.category.id
            # Convert polygon to a simple rectangle
            x1, y1, x2, y2 = xyxy
            detection_list.append([x1, y1, x2, y2, score, cls])

        return detection_list

def load_models(model_paths):
    """Load one or two YOLOv8 models."""
    models = []
    for p in model_paths:
        model = YOLOv8ModelWrapper(p, conf=CONF_THRESHOLD)
        models.append(model)
    return models

# ------------------------------
# Main Video Processing
# ------------------------------

def main():
    # Load one or two models
    models = load_models(MODEL_PATHS)

    cap = cv2.VideoCapture(VIDEO_PATH)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(OUTPUT_PATH, fourcc, fps, (width, height))

    frame_index = 0
    total_objects_detected = 0

    inference_start_time = time.time()
    processed_frames = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_index += 1

        # Run inference with one or two models
        combined_detections = []
        for model in models:
            dets = model.predict_frame(frame, use_slicing=USE_SAHI_SLICING,
                                       slice_size=SLICE_SIZE, overlap_ratio=OVERLAP_RATIO)
            # Just extend combined_detections with all results from this model
            combined_detections.extend(dets)

        # Count objects and draw bounding boxes
        frame_objects_count = 0
        for det in combined_detections:
            # det format: [x1, y1, x2, y2, conf, cls]
            x1, y1, x2, y2, conf, cls = det
            frame_objects_count += 1
            # If you want to differentiate model detections, you can color differently
            # or store model-specific info. For now, we treat all as green boxes.
            class_name = "object"
            # If using one model, we can access model names:
            # If all models have names defined, you could pick from the first model:
            # class_name = models[0].model.model.names[int(cls)] if hasattr(models[0].model.model, 'names') else str(cls)
            # If running two models, ensure they share class definitions or handle separately.

            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0,255,0), 2)
            cv2.putText(frame, f"{class_name} {conf:.2f}", (int(x1), int(y1)-5),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)

        total_objects_detected += frame_objects_count

        # Calculate current FPS
        processed_frames += 1
        elapsed_time = time.time() - inference_start_time
        current_fps = processed_frames / elapsed_time if elapsed_time > 0 else 0

        # Display FPS and object count on the frame
        cv2.putText(frame, f"FPS: {current_fps:.2f}", (10, 20),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)
        cv2.putText(frame, f"Frame Objects: {frame_objects_count}", (10, 50),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)
        # cv2.putText(frame, f"Total Objects: {total_objects_detected}", (10, 80),
        #             cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255,255,255), 2)

        out.write(frame)

        # Optional live display (comment out if not needed):
        # cv2.imshow("Frame", frame)
        # if cv2.waitKey(1) & 0xFF == ord('q'):
        #    break

    cap.release()
    out.release()
    cv2.destroyAllWindows()

    print(f"Processed {frame_index} frames.")
    print(f"Total objects detected: {total_objects_detected}")
    print(f"Average FPS: {processed_frames/(time.time() - inference_start_time):.2f}")

if __name__ == "__main__":
    main()

```

```python
# yolo tensorRT sahi
import logging
from typing import Any, Optional, List

import cv2
import numpy as np
import torch

from sahi.models.base import DetectionModel
from sahi.prediction import ObjectPrediction
from sahi.utils.compatibility import fix_full_shape_list, fix_shift_amount_list
from sahi.utils.cv import get_coco_segmentation_from_bool_mask
from sahi.utils.import_utils import check_requirements

logger = logging.getLogger(__name__)


class Yolov8EngineDetectionModel(DetectionModel):
    def check_dependencies(self) -> None:
        # –ü–µ—Ä–µ–∫–æ–Ω—É—î–º–æ—Å—å, —â–æ ultralytics –≤—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∏–π
        check_requirements(["ultralytics"])

    def load_model(self):
        """
        –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ YOLOv8 –∑ engine-—Ñ–∞–π–ª—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é ultralytics YOLO.
        """
        from ultralytics import YOLO

        try:
            model = YOLO(self.model_path)  # –ó–∞–≤–∞–Ω—Ç–∞–∂—É—î–º–æ –º–æ–¥–µ–ª—å –∑ .engine —Ñ–∞–π–ª—É
            model.to(self.device)
            self.set_model(model)
        except Exception as e:
            raise TypeError("model_path is not a valid yolov8 engine model path: ", e)

    def set_model(self, model: Any):
        """
        –í—Å—Ç–∞–Ω–æ–≤–ª—é—î –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω—É –º–æ–¥–µ–ª—å YOLOv8.
        """
        self.model = model
        # –Ø–∫—â–æ –Ω–µ –∑–∞–¥–∞–Ω–æ –º–∞–ø–ø—ñ–Ω–≥ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π, –±–µ—Ä–µ–º–æ –∑ –º–æ–¥–µ–ª—ñ
        if not self.category_mapping:
            category_mapping = {str(ind): category_name for ind, category_name in enumerate(self.category_names)}
            self.category_mapping = category_mapping

    def perform_inference(self, image: np.ndarray):
        """
        –í–∏–∫–æ–Ω—É—î –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–Ω—è –æ–±'—î–∫—Ç—ñ–≤ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–æ—ó –º–æ–¥–µ–ª—ñ YOLOv8 TensorRT.
        –ê—Ä–≥—É–º–µ–Ω—Ç–∏:
            image: np.ndarray - –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è (RGB) –¥–ª—è –¥–µ—Ç–µ–∫—Ü—ñ—ó.
        """

        from ultralytics.engine.results import Masks

        if self.model is None:
            raise ValueError("–ú–æ–¥–µ–ª—å –Ω–µ –∑–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–∞, –≤–∏–∫–ª–∏—á—Ç–µ .load_model()")

        kwargs = {"verbose": False, "conf": self.confidence_threshold, "device": self.device}
        if self.image_size is not None:
            kwargs["imgsz"] = self.image_size

        # –£–≤–∞–≥–∞: ultralytics YOLO –æ—á—ñ–∫—É—î BGR, —è–∫—â–æ –ø–µ—Ä–µ–¥–∞—î–º–æ numpy.
        prediction_result = self.model(image[:, :, ::-1], **kwargs)

        # –Ø–∫—â–æ –º–æ–¥–µ–ª—å —Å–µ–≥–º–µ–Ω—Ç–∞—Ü—ñ–π–Ω–∞
        if self.has_mask:
            if not prediction_result[0].masks:
                prediction_result[0].masks = Masks(
                    torch.tensor([], device=self.model.device), prediction_result[0].boxes.orig_shape
                )
            prediction_result = [
                (
                    result.boxes.data,
                    result.masks.data,
                )
                for result in prediction_result
            ]
        else:
            prediction_result = [result.boxes.data for result in prediction_result]

        self._original_predictions = prediction_result
        self._original_shape = image.shape

    @property
    def category_names(self):
        return self.model.names.values()

    @property
    def num_categories(self):
        return len(self.model.names)

    @property
    def has_mask(self):
        return self.model.overrides["task"] == "segment"

    def _create_object_prediction_list_from_original_predictions(
        self,
        shift_amount_list: Optional[List[List[int]]] = [[0, 0]],
        full_shape_list: Optional[List[List[int]]] = None,
    ):
        """
        –ö–æ–Ω–≤–µ—Ä—Ç—É—î –æ—Ä–∏–≥—ñ–Ω–∞–ª—å–Ω—ñ –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ –≤ —Å–ø–∏—Å–æ–∫ –æ–±'—î–∫—Ç—ñ–≤ SAHI ObjectPrediction.
        """
        original_predictions = self._original_predictions

        shift_amount_list = fix_shift_amount_list(shift_amount_list)
        full_shape_list = fix_full_shape_list(full_shape_list)

        object_prediction_list_per_image = []
        for image_ind, image_predictions in enumerate(original_predictions):
            shift_amount = shift_amount_list[image_ind]
            full_shape = None if full_shape_list is None else full_shape_list[image_ind]
            object_prediction_list = []

            if self.has_mask:
                image_predictions_in_xyxy_format = image_predictions[0]
                image_predictions_masks = image_predictions[1]
                for prediction, bool_mask in zip(
                    image_predictions_in_xyxy_format.cpu().detach().numpy(),
                    image_predictions_masks.cpu().detach().numpy(),
                ):
                    x1, y1, x2, y2, score, category_id = prediction
                    bbox = [x1, y1, x2, y2]
                    category_id = int(category_id)
                    category_name = self.category_mapping[str(category_id)]

                    orig_width = self._original_shape[1]
                    orig_height = self._original_shape[0]
                    bool_mask = cv2.resize(bool_mask.astype(np.uint8), (orig_width, orig_height))
                    segmentation = get_coco_segmentation_from_bool_mask(bool_mask)
                    if len(segmentation) == 0:
                        continue

                    # –í–∏–ø—Ä–∞–≤–ª—è—î–º–æ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ñ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏
                    bbox[0] = max(0, bbox[0])
                    bbox[1] = max(0, bbox[1])
                    bbox[2] = max(0, bbox[2])
                    bbox[3] = max(0, bbox[3])

                    # –í–∏–ø—Ä–∞–≤–ª—è—î–º–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏ –∑–∞ –º–µ–∂–∞–º–∏ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è
                    if full_shape is not None:
                        bbox[0] = min(full_shape[1], bbox[0])
                        bbox[1] = min(full_shape[0], bbox[1])
                        bbox[2] = min(full_shape[1], bbox[2])
                        bbox[3] = min(full_shape[0], bbox[3])

                    if not (bbox[0] < bbox[2]) or not (bbox[1] < bbox[3]):
                        logger.warning(f"ignoring invalid prediction with bbox: {bbox}")
                        continue

                    object_prediction = ObjectPrediction(
                        bbox=bbox,
                        category_id=category_id,
                        score=score,
                        segmentation=segmentation,
                        category_name=category_name,
                        shift_amount=shift_amount,
                        full_shape=full_shape,
                    )
                    object_prediction_list.append(object_prediction)
                object_prediction_list_per_image.append(object_prediction_list)
            else:
                # –¢—ñ–ª—å–∫–∏ –±–æ–∫—Å–∏
                for prediction in image_predictions.cpu().detach().numpy():
                    x1, y1, x2, y2, score, category_id = prediction
                    bbox = [x1, y1, x2, y2]
                    category_id = int(category_id)
                    category_name = self.category_mapping[str(category_id)]

                    # –í–∏–ø—Ä–∞–≤–ª–µ–Ω–Ω—è –Ω–µ–≥–∞—Ç–∏–≤–Ω–∏—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç
                    bbox[0] = max(0, bbox[0])
                    bbox[1] = max(0, bbox[1])
                    bbox[2] = max(0, bbox[2])
                    bbox[3] = max(0, bbox[3])

                    if full_shape is not None:
                        bbox[0] = min(full_shape[1], bbox[0])
                        bbox[1] = min(full_shape[0], bbox[1])
                        bbox[2] = min(full_shape[1], bbox[2])
                        bbox[3] = min(full_shape[0], bbox[3])

                    if not (bbox[0] < bbox[2]) or not (bbox[1] < bbox[3]):
                        logger.warning(f"ignoring invalid prediction with bbox: {bbox}")
                        continue

                    object_prediction = ObjectPrediction(
                        bbox=bbox,
                        category_id=category_id,
                        score=score,
                        segmentation=None,
                        category_name=category_name,
                        shift_amount=shift_amount,
                        full_shape=full_shape,
                    )
                    object_prediction_list.append(object_prediction)
                object_prediction_list_per_image.append(object_prediction_list)

        self._object_prediction_list_per_image = object_prediction_list_per_image
```

```python
import cv2
import numpy as np
# –ö–ª–∞—Å, –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑–æ–≤–∞–Ω–∏–π –≤–∏—â–µ
from ultralytics import YOLO

# –®–ª—è—Ö –¥–æ .engine –º–æ–¥–µ–ª—ñ
MODEL_PATH = "/content/yolov8n.engine"

# –ü–∞—Ä–∞–º–µ—Ç—Ä–∏ –º–æ–¥–µ–ª—ñ
CONFIDENCE_THRESHOLD = 0.5
IMAGE_SIZE = 640
DEVICE = "cuda"  # –∞–±–æ "cpu"

# –Ü–Ω—ñ—Ü—ñ–∞–ª—ñ–∑–∞—Ü—ñ—è YOLOv8 –º–æ–¥–µ–ª—ñ
model = Yolov8EngineDetectionModel(
    model_path=MODEL_PATH,
    confidence_threshold=CONFIDENCE_THRESHOLD,
    image_size=IMAGE_SIZE,
    device=DEVICE,
)
model.load_model()

model = YOLO('/content/yolov8n.engine')
results = model.predict(source='path/to/your/image.jpg', device=0)  # –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –ø–∞—Ä–∞–º–µ—Ç—Ä device —Ç—É—Ç


# –í—ñ–¥–µ–æ –≤—Ö—ñ–¥ —ñ –≤–∏—Ö—ñ–¥
VIDEO_INPUT_PATH = "/content/Cars1.mp4"
VIDEO_OUTPUT_PATH = "/content/output_video.mp4"

# –í—ñ–¥–∫—Ä–∏—Ç—Ç—è –≤—ñ–¥–µ–æ
cap = cv2.VideoCapture(VIDEO_INPUT_PATH)
if not cap.isOpened():
    print("–ü–æ–º–∏–ª–∫–∞: –ù–µ–º–æ–∂–ª–∏–≤–æ –≤—ñ–¥–∫—Ä–∏—Ç–∏ –≤—ñ–¥–µ–æ.")
    exit()

# –ù–∞–ª–∞—à—Ç—É–≤–∞–Ω–Ω—è –≤—ñ–¥–µ–æ –¥–ª—è –∑–∞–ø–∏—Å—É
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
out = cv2.VideoWriter(
    VIDEO_OUTPUT_PATH,
    cv2.VideoWriter_fourcc(*'mp4v'),
    fps,
    (frame_width, frame_height),
)

# –û–±—Ä–æ–±–∫–∞ –∫–æ–∂–Ω–æ–≥–æ –∫–∞–¥—Ä—É
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # –ó–∞–∫—ñ–Ω—á–µ–Ω–Ω—è –≤—ñ–¥–µ–æ

    # –í–∏–∫–æ–Ω–∞–Ω–Ω—è –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω–Ω—è
    model.perform_inference(frame)
    predictions = model._create_object_prediction_list_from_original_predictions()

    # –í—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è –ø–µ—Ä–µ–¥–±–∞—á–µ–Ω—å –Ω–∞ –∫–∞–¥—Ä—ñ
    for obj in predictions[0]:  # –û–±—Ä–æ–±–ª—è—î–º–æ –ª–∏—à–µ –ø–µ—Ä—à–∏–π –∫–∞–¥—Ä (batch-size=1)
        bbox = obj.bbox
        category_name = obj.category_name
        score = obj.score

        # –í–∏–≤–æ–¥–∏–º–æ —Ä–∞–º–∫—É —Ç–∞ –ø—ñ–¥–ø–∏—Å
        cv2.rectangle(frame, (int(bbox[0]), int(bbox[1])), (int(bbox[2]), int(bbox[3])), (0, 255, 0), 2)
        text = f"{category_name}: {score:.2f}"
        cv2.putText(
            frame,
            text,
            (int(bbox[0]), int(bbox[1]) - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.5,
            (0, 255, 0),
            2,
        )

    # –ó–∞–ø–∏—Å –æ–±—Ä–æ–±–ª–µ–Ω–æ–≥–æ –∫–∞–¥—Ä—É
    out.write(frame)

    # –î–ª—è –ø–µ—Ä–µ–≥–ª—è–¥—É –≤ —Ä–µ–∞–ª—å–Ω–æ–º—É —á–∞—Å—ñ
    cv2.imshow("Object Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# –ó–∞–≤–µ—Ä—à–µ–Ω–Ω—è
cap.release()
out.release()
cv2.destroyAllWindows()
print("–û–±—Ä–æ–±–∫–∞ –≤—ñ–¥–µ–æ –∑–∞–≤–µ—Ä—à–µ–Ω–∞.")

```

```python
import cv2
import os
import torch
import numpy as np
import random
import time
import pathlib
from ultralytics.utils import ops
from ultralytics.data.augment import LetterBox
from ultralytics.engine.results import Results
from ultralytics import YOLO

def get_text_color(box_color):
    brightness = box_color[2]*0.299 + box_color[1]*0.587 + box_color[0]*0.114
    return (0,0,0) if brightness > 180 else (255,255,255)

def draw_fps(avg_fps, combined_img):
    avg_fps_str = float("{:.2f}".format(avg_fps))
    cv2.putText(combined_img, "FPS: "+str(avg_fps_str), (5, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), thickness=3)
    return combined_img

def get_name(source_path):
    file_pos = (source_path).rfind('\\')
    if(file_pos == -1):
        file_pos = (source_path).rfind('/')
        if(file_pos == -1):
            file_pos = 0
    name_idx = file_pos + 1
    name = source_path[name_idx:]
    return name

def get_save_path(file_name, folder_name):
    path = "result"
    save_path = os.path.join(path, folder_name)
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    save_path = os.path.join(save_path, file_name)
    return save_path

def preprocess(img):
    # LetterBox
    im = LetterBox((640, 640), False)(image=img)
    im = im.transpose((2, 0, 1))[::-1]  # HWC to CHW, BGR to RGB
    im = np.ascontiguousarray(im)
    img = torch.from_numpy(im).to(torch.device('cuda:0'))
    img = img.half()
    img /= 255
    if len(img.shape) == 3:
        img = img[None]
    return img

# def preprocess(img, size=(320, 320)):
#     # –ó–º—ñ–Ω–µ–Ω–æ —Ä–æ–∑–º—ñ—Ä –Ω–∞ 320x320
#     im = LetterBox(size, False)(image=img)
#     im = im.transpose((2, 0, 1))[::-1]  # HWC to CHW, BGR to RGB
#     im = np.ascontiguousarray(im)
#     img = torch.from_numpy(im).to(torch.device('cuda:0'))
#     img = img.half()  # –ü–µ—Ä–µ—Ç–≤–æ—Ä–µ–Ω–Ω—è –Ω–∞ FP16
#     img /= 255
#     if len(img.shape) == 3:
#         img = img[None]
#     return img

# def check_and_resize(img, size=(320, 320)):
#     if img.shape[2:] != size:
#         img = preprocess(img, size)
#     return img


def postprocess(preds, img, orig_img, names, source_path):
    preds = ops.non_max_suppression(preds,
                                    0.5,
                                    0.7,
                                    agnostic=False,
                                    max_det=300,
                                    classes=None)

    results = []
    for i, pred in enumerate(preds):
        orig_img_ = orig_img[i] if isinstance(orig_img, list) else orig_img
        shape = orig_img_.shape
        pred[:, :4] = ops.scale_boxes(img.shape[2:], pred[:, :4], shape).round()
        path = source_path
        img_path = path[i] if isinstance(path, list) else path
        results.append(Results(orig_img=orig_img_, path=img_path, names=names, boxes=pred))
    return results

def frame_objects_count(frame, detections, class_list, colors):
    frame_objects_count_val = 0
    for cls, conf, box in detections:
        frame_objects_count_val += 1
        x1, y1, x2, y2 = box
        label = class_list[int(cls)]
        box_color = colors[int(cls)]
        text_color = get_text_color(box_color)
        # Draw bounding box
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), box_color, 1)
        label_with_conf = f"{label} {conf:.2f}"
        text_size, _ = cv2.getTextSize(label_with_conf, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)
        text_width, text_height = text_size
        # cv2.rectangle(frame, (int(x1), int(y1) - text_height - 4),
        #               (int(x1) + text_width, int(y1)), box_color, -1)
        cv2.putText(frame, label_with_conf, (int(x1), int(y1) - 2),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, text_color, 1)
    cv2.putText(frame, f"Frame Objects: {frame_objects_count_val}", (10, 50),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
    return frame, frame_objects_count_val

############################
# –ù–æ–≤—ñ –¥–æ–ø–æ–º—ñ–∂–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—ó
############################

def slice_image(image, tile_size=640, overlap=0.2):
    """
    –†–æ–∑–±–∏–≤–∞—î –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è –Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ñ —Ñ—Ä–∞–≥–º–µ–Ω—Ç–∏ (tile) –¥–ª—è –¥–µ—Ç–µ–∫—Ü—ñ—ó.
    tile_size: —Ä–æ–∑–º—ñ—Ä —Ñ—Ä–∞–≥–º–µ–Ω—Ç—É (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥ 640x640)
    overlap: –ø–µ—Ä–µ–∫—Ä–∏—Ç—Ç—è –º—ñ–∂ —Ñ—Ä–∞–≥–º–µ–Ω—Ç–∞–º–∏ —É —á–∞—Å—Ç–∫–∞—Ö –≤—ñ–¥ tile_size
    –ü–æ–≤–µ—Ä—Ç–∞—î —Å–ø–∏—Å–æ–∫ —Ñ—Ä–∞–≥–º–µ–Ω—Ç—ñ–≤ —Ç–∞ —ó—Ö –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∏ —É –≤–∏—Ö—ñ–¥–Ω–æ–º—É –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—ñ.
    """
    h, w, _ = image.shape
    step = int(tile_size * (1 - overlap))
    tiles = []
    coords = []

    for y in range(0, h, step):
        for x in range(0, w, step):
            y_end = min(y + tile_size, h)
            x_end = min(x + tile_size, w)
            tile = image[y:y_end, x:x_end]
            tiles.append(tile)
            coords.append((x, y, x_end, y_end))
    return tiles, coords

def detect_on_tile(model, tile):
    """
    –í–∏–∫–æ–Ω—É—î –¥–µ—Ç–µ–∫—Ü—ñ—é –Ω–∞ –æ–∫—Ä–µ–º–æ–º—É —Ñ—Ä–∞–≥–º–µ–Ω—Ç—ñ –∑–æ–±—Ä–∞–∂–µ–Ω–Ω—è
    –ü–æ–≤–µ—Ä—Ç–∞—î (cls, conf, box) –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –¥–µ—Ç–µ–∫—Ç–æ–≤–∞–Ω–æ–≥–æ –æ–±'—î–∫—Ç—É.
    """
    results = model.predict(tile, imgsz=640, conf=0.5, verbose=False)
    result = results[0].cpu()
    box = result.boxes.xyxy.numpy()
    conf = result.boxes.conf.numpy()
    cls = result.boxes.cls.numpy().astype(int)
    return list(zip(cls, conf, box))

def merge_detections(detections_list, coords_list):
    all_detections = []
    for dets, (x0, y0, x1, y1) in zip(detections_list, coords_list):
        for cls, conf, box in dets:
            bx1, by1, bx2, by2 = box
            # –ü–µ—Ä–µ–≤–µ–¥–µ–Ω–Ω—è –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —É –≥–ª–æ–±–∞–ª—å–Ω—É —Å–∏—Å—Ç–µ–º—É
            bx1_global = bx1 + x0
            by1_global = by1 + y0
            bx2_global = bx2 + x0
            by2_global = by2 + y0
            all_detections.append((cls, conf, [bx1_global, by1_global, bx2_global, by2_global]))

    # –Ø–∫—â–æ –Ω–µ–º–∞—î –∂–æ–¥–Ω–∏—Ö –¥–µ—Ç–µ–∫—Ü—ñ–π, –ø–æ–≤–µ—Ä—Ç–∞—î–º–æ –ø–æ—Ä–æ–∂–Ω—ñ–π —Å–ø–∏—Å–æ–∫
    if len(all_detections) == 0:
        return []

    all_boxes = []
    all_confs = []
    all_clss = []
    for (cls, conf, box) in all_detections:
        all_clss.append(cls)
        all_confs.append(conf)
        all_boxes.append(box)

    boxes_tensor = torch.tensor(all_boxes, dtype=torch.float32)
    confs_tensor = torch.tensor(all_confs, dtype=torch.float32).unsqueeze(1)
    clss_tensor = torch.tensor(all_clss, dtype=torch.float32).unsqueeze(1)
    pred = torch.cat([boxes_tensor, confs_tensor, clss_tensor], dim=1).unsqueeze(0)

    # –Ø–∫—â–æ pred –ø–æ—Ä–æ–∂–Ω—ñ–π (—Ç–æ–±—Ç–æ pred.shape[1] == 0), –ø–æ–≤–µ—Ä—Ç–∞—î–º–æ –ø—É—Å—Ç–∏–π —Å–ø–∏—Å–æ–∫
    if pred.shape[1] == 0:
        return []

    # –í–∏–∫–æ–Ω—É—î–º–æ NMS –ª–∏—à–µ —è–∫—â–æ —î –¥–µ—Ç–µ–∫—Ü—ñ—ó
    nms_output = ops.non_max_suppression(pred, 0.5, 0.7, agnostic=False, max_det=300, classes=None)
    final_detections = []
    for det in nms_output[0]:
        x1, y1, x2, y2, cf, c = det
        final_detections.append((int(c.item()), cf.item(), [x1.item(), y1.item(), x2.item(), y2.item()]))

    return final_detections


def get_detection_result(model, frame):
    # –¶—è —Ñ—É–Ω–∫—Ü—ñ—è –Ω–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î—Ç—å—Å—è –Ω–∞–ø—Ä—è–º—É, –±–æ –º–∏ –±—É–¥–µ–º–æ –æ–±—Ä–æ–±–ª—è—Ç–∏ tile
    results = model.predict(frame, imgsz=640, conf=0.5, verbose=False)
    result = results[0].cpu()
    box = result.boxes.xyxy.numpy()
    conf = result.boxes.conf.numpy()
    cls = result.boxes.cls.numpy().astype(int)
    return cls, conf, box

def detection(model_path, source, name, tile_size=640, overlap=0.2):
    # –ó–∞–≤–∞–Ω—Ç–∞–∂–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ
    model = YOLO(model_path, task='detect')
    label_map = model.names
    COLORS = [[random.randint(0, 255) for _ in range(3)] for _ in label_map]

    frame_count = 0
    total_fps = 0
    avg_fps = 0

    video_cap = cv2.VideoCapture(source)
    total_frames = int(video_cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_width = int(video_cap.get(3))
    frame_height = int(video_cap.get(4))

    video_frames = []

    while video_cap.isOpened():
        ret, frame = video_cap.read()
        if not ret:
            break

        start = time.time()

        # –†–æ–∑–±–∏–≤–∞—î–º–æ –∫–∞–¥—Ä –Ω–∞ tile
        tiles, coords = slice_image(frame, tile_size=tile_size, overlap=overlap)

        # –î–µ—Ç–µ–∫—Ü—ñ—è –Ω–∞ –∫–æ–∂–Ω–æ–º—É tile
        tile_detections = []
        for timg in tiles:
            dets = detect_on_tile(model, timg)
            tile_detections.append(dets)

        # –ó–±–∏—Ä–∞—î–º–æ —Ç–∞ –æ–±'—î–¥–Ω—É—î–º–æ –¥–µ—Ç–µ–∫—Ü—ñ—ó
        final_dets = merge_detections(tile_detections, coords)

        # –ú–∞–ª—é—î–º–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∏
        frame, frame_objects_count_value = frame_objects_count(frame, final_dets, label_map, COLORS)

        end = time.time()
        frame_count += 1
        fps = 1 / (end - start)
        total_fps += fps
        avg_fps = total_fps / frame_count

        frame = draw_fps(avg_fps, frame)

        video_frames.append(frame)

        print(f"Frame {frame_count}/{total_frames} processed. Objects detected: {frame_objects_count_value}")

    # –ó–±–µ—Ä–µ–∂–µ–Ω–Ω—è —Ä–µ–∑—É–ª—å—Ç—É—é—á–æ–≥–æ –≤—ñ–¥–µ–æ
    file_name = get_name(source)
    save_path = get_save_path(file_name, name)
    out = cv2.VideoWriter(save_path, cv2.VideoWriter_fourcc(*'XVID'), int(avg_fps), (frame_width, frame_height))

    for f in video_frames:
        out.write(f)

    out.release()
    print(f"Video is saved in: {save_path}")

# –ü—Ä–∏–∫–ª–∞–¥ –≤–∏–∫–ª–∏–∫—É:
# detection("yolov8n.engine", "input_video.mp4", "sliced_result", tile_size=640, overlap=0.2)
```
